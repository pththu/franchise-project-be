```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import json
import time
```

```python
base_url = "https://www.highlandscoffee.com.vn/vn/"
URLS = ["highlands-teas", "freeze"]

chrome_options = Options()
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_argument("--start-maximized")
driver = webdriver.Chrome(options=chrome_options)

for name in URLS:
    url = f"{base_url}{name}.html"
    driver.get(url)
    time.sleep(10)

    html = driver.page_source

    with open(f"page_{name}.html", "w", encoding="utf-8") as f:
        f.write(html)
        
print("Crawl Done.")
driver.quit()
```

```python
seen = set() 
products = []
id = 0 

for name_url in URLS:
    with open(f"page_{name_url}.html", "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "lxml")

    for item in soup.select("div.product"):
        
        name_tag = item.select_one(".tend2 h3")
        price_tag = item.select_one(".price strong")
        img_tag = item.select_one(".img img")
        
        if not (name_tag and price_tag and img_tag):
            continue
        
        name = name_tag.get_text(strip=True)
        price = price_tag.get_text(strip=True)
        img = img_tag.get("src")
        
        if name in seen:
            continue
        seen.add(name)
        
        products.append({
            "id": id,
            "name": name,
            "price": price,
            "image": img
        })

        id += 1


with open(f"products_drinks.json", "w", encoding="utf-8") as f:
    json.dump(products, f, ensure_ascii=False, indent=4)
print(f"Đã lọc các sản phẩm duy nhất file chứa {id + 1} sản phẩm")
```
